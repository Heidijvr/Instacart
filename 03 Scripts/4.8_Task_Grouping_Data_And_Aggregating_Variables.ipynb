## 5. Spending flag for each user based on the average price across all their orders 

```python
# Create mean_price_ords column based on user_id:
ords_prods_merge['mean_price_ords'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python

# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
ords_prods_merge.loc[ords_prods_merge['mean_price_ords'] < 10, 'spending_flag'] = 'Low spender'


# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
ords_prods_merge.loc[ords_prods_merge['mean_price_ords'] >= 10, 'spending_flag'] = 'High spender'

```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

Deduction: There are roughly far less more Higher spenders than Low spenders.

## 6. Order-frequency flag based on 'days_since_prior_order'

```python
# In order to send relevant notifications to users within the app (for instance, asking users if they want to buy 
# the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order 
# frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order”  
# column. The criteria for the flag should be as follows:
```

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# a. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# b. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# c. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.
```

```python
# Create categories based on these order frequencies
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'customer_frequency_flag'] = 'Non_Frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'customer_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'customer_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['customer_frequency_flag'].value_counts(dropna = False)
```
