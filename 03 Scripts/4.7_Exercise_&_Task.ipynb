## 3. Categorizing Products by Price_range using User-defined Functions

```python
#Create a subset of first 1000000 rows
df = ords_prods_merged[:1000000]
```

```python
df.head()
```

```python
df.shape
```

```python
# Define price function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row ['prices'] > 15:
        return 'High-range'
    else: return 'Not enough data'
```

```python
# Imlementing new column title: price_range

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in new column
df['price_range'].value_counts(dropna = False)
```

```python
#Check for the most expensive product in the subset
df['prices'].max()
```

```python
df['price_range'].describe()
```

## 4. Categorizing Products by Price-range using loc() on Subset 

```python
#Using the loc(function) to apply the conditional logic of an if-statement
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Checking values in new column
df['price_range_loc'].value_counts(dropna = False)
```

## 6.1 Updating Busiest Day Column

```python
# Frequency on 'orders_day_of_the_week' column. What items are being bought on busiest and slowest days
```

```python
df['orders_day_of_the_week'].value_counts(dropna = False)
```

```python
#Deduction: 0 = Sunday, has the highest frequency; 4 - Thursday has the lowest frequency
```

```python
# Create new columns for “Busiest day,” “Least busy,” and “Regularly busy” using for-loop
```

```python
result = []

for value in ords_prods_merged['orders_day_of_the_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regular busy')
```

```python
result
```

```python
# Create new column within df_ords_prods_merged and set it equal to result
ords_prods_merged['busiest_day'] = result
```

```python
# Frequency of this new column
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

Change “Busiest day” to “Busiest days” and slowest days to two slowest days (plural). 
Create new columns for this using a suitable method.

```python
# Create new columns for “Two Busiest days,” “Two Least busiest,” and “Regular busy days.”
result_1 = []

for value in ords_prods_merged["orders_day_of_the_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 3 or value == 4:
    result_1.append("Least busy")
  else:
    result_1.append("Regularly busy")
        
```

```python
ords_prods_merged['busiest_days'] = result_1
```

```python
# Check the values of this new column for accuracy
ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
crosstab = pd.crosstab(ords_prods_merged['orders_day_of_the_week'], ords_prods_merged['busiest_days'], dropna = False)
```

```python
# Copy crosstab to clipboard
crosstab.to_clipboard()
```

```python
crosstab
```

```python
# Observations: Looking at the value counts in the 'order_dow'; By adding up 0 and 1, the result accurately reflects the results
# of the TWO Busiest days. This observation is also true for adding up the results of 3 and 4 of the value counts in the 
# 'order_dow' and the reflection of the results of the TWO Least busiest days. 
```

## Question 4

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer 
at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled 
“Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called 
“busiest_period_of_day.”

```python
# Identify frequency of order hours
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
ords_prods_merged.head()
```

```python
# Deduction: 10:00am is the MOST popular time to order and 3:00am the LEAST popular
```

```python
# Create new columns for “Most orders,” “Average orders,” and “Fewest orders.”
```

```python
result = []

for value in ords_prods_merged['order_hour_of_day']:
    if value >= 0 and value <= 6:
        result.append('Fewest orders')
    elif value >= 9 and value <= 17:
        result.append('Most orders')
    else: result.append('Average orders') 
```

```python
# Create column 'busiest_period_of_day'
ords_prods_merged['busiest_period_of_day'] = result
```

```python
# Print frequency of new column
ords_prods_merged['busiest_period_of_day'].value_counts (dropna = False)
```

```python
crosstab = pd.crosstab(ords_prods_merged['order_hour_of_day'], ords_prods_merged['busiest_period_of_day'], dropna = False)
```

```python
# Copy crosstab to clipboard
crosstab.to_clipboard()
```
